```python
import lightkurve as lk
import pandas as pd

targets = pd.read_csv("data/targets/target_list.csv")
sample_ids = targets["TIC_ID"].astype(str).tolist()[:30]  # test first 30

sector_counts = {}
for tic in sample_ids:
    search = lk.search_lightcurve(f"TIC {tic}", mission="TESS", author="SPOC", cadence="short")
    for row in search:
        sector = row.mission[0]  # e.g. "TESS Sector 12"
        sector_counts[sector] = sector_counts.get(sector, 0) + 1

for sector, count in sorted(sector_counts.items(), key=lambda x: -x[1]):
    print(f"{sector}: {count} stars have 2-min SPOC data")
```

```python
import os

raw_12 = set(f.replace("TIC_", "").replace(".fits", "") for f in os.listdir("data/raw/sector_12") if f.endswith(".fits"))
raw_13 = set(f.replace("TIC_", "").replace(".fits", "") for f in os.listdir("data/raw/sector_13") if f.endswith(".fits"))
all_raw = raw_12 | raw_13

processed = set(f.replace("TIC_", "").replace(".npy", "") for f in os.listdir("data/processed") if f.endswith(".npy"))

missing = all_raw - processed
print(f"Raw FITS files: {len(all_raw)}")
print(f"Processed files: {len(processed)}")
print(f"Missing (failed during preprocessing): {len(missing)}")
print(missing)
```
